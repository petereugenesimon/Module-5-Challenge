### Analysis

- When first looking through the summary table, it would appear that Capomulin and Ramicane are the most successful drugs based on how much lower their average tumor volume is compared to the other drugs. Both of them showing almost the same average, sitting around 40.2-40.6, while the remainder of the drugs are sit in the 50s.

- That first observation could be due to the fact that, looking at the bar charts, that almost 50 more mice were tested using Capomulin and Ramicane. Although that may not be the reason for the results, it is something that should be taken into account considering that's almost 25% more mice than the other drugs used.

- When looking at how long it takes for the drug to "kick in," by looking at the line chart, it could be deduced that it takes about 3 weeks before the drug starts to work it's course and mitigate the tumor. We would have to plot more mice though to come up with a more definitive answer as it the drug may affect other mice differently.

- Finally, we can confidently say that there is a strong correlation between the average tumor volume and the weight of a mouse. Contextually-speaking, this makes sense because when a mouse is larger, there are more cells that could potentially be affected by the tumor.
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
study_df = pd.merge(study_results, mouse_metadata, on=["Mouse ID"])

# Display the data table for preview
study_df.head()
```

```python
# Checking the number of mice.
n_mice = len(pd.unique(study_df['Mouse ID']))
n_mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicates = study_df[study_df.duplicated(['Mouse ID', 'Timepoint'])]
print(duplicates)
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
dup_study_df = pd.DataFrame(study_df.loc[study_df['Mouse ID']=='g989',:])
dup_study_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
study_df = study_df.loc[study_df["Mouse ID"] !="g989"]
study_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
n_mice = len(pd.unique(study_df['Mouse ID']))
n_mice
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

study_df_grp = study_df.groupby('Drug Regimen')

mean_tumor_volume = study_df_grp["Tumor Volume (mm3)"].mean()
med_tumor_volume = study_df_grp["Tumor Volume (mm3)"].median()
var_tumor_volume = study_df_grp["Tumor Volume (mm3)"].var()
std_tumor_volume = study_df_grp["Tumor Volume (mm3)"].std()
sem_tumor_volume = study_df_grp["Tumor Volume (mm3)"].sem()

stat_summary_df = pd.DataFrame({"Mean Tumor Volume": mean_tumor_volume,
                                "Median Tumor Volume": med_tumor_volume,
                                "Tumor Volume Variance": var_tumor_volume,
                               "Tumor Volume Std. Dev.": std_tumor_volume,
                               "Tumor Volume Std. Err.": sem_tumor_volume})
stat_summary_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line.
study_df_grp.agg({'Tumor Volume (mm3)': ['mean', 'median', 'var', 'std', 'sem']})
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse


# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
```

```python
mouse_id_grp = study_df.groupby('Mouse ID')

last_timepoint_df = (pd.concat([mouse_id_grp.tail(1)])
                        .drop_duplicates()
                        .sort_values('Mouse ID')
                        .reset_index(drop=True))

drug_regimens = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

drug_reg_df = last_timepoint_df[last_timepoint_df['Drug Regimen'].isin(drug_regimens)]
```

```python
# Put treatments into a list for for loop (and later for plot labels)


# Create empty list to fill with tumor vol data (for plotting)


# Calculate the IQR and quantitatively determine if there are any potential outliers. 

    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    
    
    # add subset 
    
    
    # Determine outliers using upper and lower bounds
```

```python
cap_tum_vol = []
ram_tum_vol = []
inf_tum_vol = []
cef_tum_vol = []

cap_df = drug_reg_df[drug_reg_df['Drug Regimen']=='Capomulin']
ram_df = drug_reg_df[drug_reg_df['Drug Regimen']=='Ramicane']
inf_df = drug_reg_df[drug_reg_df['Drug Regimen']=='Infubinol']
cef_df = drug_reg_df[drug_reg_df['Drug Regimen']=='Ceftamin']

for cap_index, cap_row in cap_df.iterrows():
    cap_tum_vol.append(cap_row['Tumor Volume (mm3)'])
    
for ram_index, ram_row in ram_df.iterrows():
    ram_tum_vol.append(ram_row['Tumor Volume (mm3)'])
    
for inf_index, inf_row in inf_df.iterrows():
    inf_tum_vol.append(inf_row['Tumor Volume (mm3)'])
    
for cef_index, cef_row in cef_df.iterrows():
    cef_tum_vol.append(cef_row['Tumor Volume (mm3)'])
    
cap_quartiles = cap_df['Tumor Volume (mm3)'].quantile([.25,.5,.75])
cap_lowerq = cap_quartiles[0.25]
cap_upperq = cap_quartiles[0.75]
cap_iqr = cap_upperq - cap_lowerq

cap_lower_bound = cap_lowerq - (1.5*cap_iqr)
cap_upper_bound = cap_upperq + (1.5*cap_iqr)

cap_outlier = cap_df.loc[(cap_df['Tumor Volume (mm3)'] < cap_lower_bound) | (cap_df['Tumor Volume (mm3)'] > cap_upper_bound)]

ram_quartiles = ram_df['Tumor Volume (mm3)'].quantile([.25,.5,.75])
ram_lowerq = ram_quartiles[0.25]
ram_upperq = ram_quartiles[0.75]
ram_iqr = ram_upperq - ram_lowerq

ram_lower_bound = ram_lowerq - (1.5*ram_iqr)
ram_upper_bound = ram_upperq + (1.5*ram_iqr)

ram_outlier = ram_df.loc[(ram_df['Tumor Volume (mm3)'] < ram_lower_bound) | (ram_df['Tumor Volume (mm3)'] > ram_upper_bound)]

inf_quartiles = inf_df['Tumor Volume (mm3)'].quantile([.25,.5,.75])
inf_lowerq = inf_quartiles[0.25]
inf_upperq = inf_quartiles[0.75]
inf_iqr = inf_upperq - inf_lowerq

inf_lower_bound = inf_lowerq - (1.5*inf_iqr)
inf_upper_bound = inf_upperq + (1.5*inf_iqr)

inf_outlier = inf_df.loc[(inf_df['Tumor Volume (mm3)'] < inf_lower_bound) | (inf_df['Tumor Volume (mm3)'] > inf_upper_bound)]

cef_quartiles = cef_df['Tumor Volume (mm3)'].quantile([.25,.5,.75])
cef_lowerq = cef_quartiles[0.25]
cef_upperq = cef_quartiles[0.75]
cef_iqr = cef_upperq - cef_lowerq

cef_lower_bound = cef_lowerq - (1.5*cef_iqr)
cef_upper_bound = cef_upperq + (1.5*cef_iqr)

cef_outlier = cef_df.loc[(cef_df['Tumor Volume (mm3)'] < cef_lower_bound) | (cef_df['Tumor Volume (mm3)'] > cef_upper_bound)]

print(f"Capomulin's potential outliers: {cap_outlier['Tumor Volume (mm3)']}")
print(f"Ramicane's potential outliers: {ram_outlier['Tumor Volume (mm3)']}")
print(f"Infubinol's potential outliers: {inf_outlier['Tumor Volume (mm3)']}")
print(f"Ceftamin's potential outliers: {cef_outlier['Tumor Volume (mm3)']}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.

boxplot_data = pd.DataFrame({'Capomulin': cap_tum_vol, 'Ramicane': ram_tum_vol, 'Infubinol': inf_tum_vol, 'Ceftamin': cef_tum_vol})
outlier_c = dict(markerfacecolor='red',marker='o')
boxplot = plt.boxplot(np.array(boxplot_data), labels=list(boxplot_data), flierprops=outlier_c)
plt.ylabel('Final Tumor Volume (mm3)')

plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
```

```python
mouse_l509_df = study_df.loc[study_df["Mouse ID"] =='l509']
l509_timepoints = mouse_l509_df["Timepoint"].values.tolist()
l509_tum_vol = mouse_l509_df["Tumor Volume (mm3)"].values.tolist()

plt.plot(l509_timepoints, l509_tum_vol)
plt.title("Capomulin Treatment of Mouse L509")
plt.xlabel("Timepoint (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
```

```python
capomulin_df = study_df.loc[study_df["Drug Regimen"] =='Capomulin']

cap_drug_grp = capomulin_df.groupby('Mouse ID')

cap_mean_tum_vol = cap_drug_grp["Tumor Volume (mm3)"].mean()
mouse_weight = cap_drug_grp["Weight (g)"].mean()

scatter_df = pd.DataFrame({"Mean Tumor Volume": cap_mean_tum_vol,
                                "Mouse Weight": mouse_weight})

cap_tum_vol_scat = scatter_df["Mean Tumor Volume"].values.tolist()
cap_weight_scat = scatter_df["Mouse Weight"].values.tolist()

plt.scatter(cap_weight_scat, cap_tum_vol_scat)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
from scipy.stats import linregress
avg_tum_arr = np.array(cap_tum_vol_scat)
weight_arr = np.array(cap_weight_scat)

(slope, intercept, rvalue, pvalue, stderr) = linregress(weight_arr, avg_tum_arr)
regress_values = weight_arr * slope + intercept
plt.scatter(weight_arr, avg_tum_arr)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.plot(weight_arr, regress_values, "r-")
print(f"The correlation between mouse weight and the average tumor volume is {rvalue:.2f}")

plt.show()
```
